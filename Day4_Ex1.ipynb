```python
import matplotlib
import matplotlib.pyplot as plt

import numpy as np

import skimage
from skimage import data
from skimage import exposure
```

```python
camera = data.camera()
```

```python
print(camera.shape)
```

```python
print(camera.max())
```

```python
print(camera.min())
```

```python
print(camera.mean())
```

```python
w,h = camera.shape
```

```python
hist = [0]*256
```

```python
for y in range(h):
        for x in range(w):
            gray_level= camera[x, y]
            hist[gray_level] = hist[gray_level]+1
```

```python
plt.plot(hist)
```

```python
plt.hist(camera.ravel(),256,[0,256]); plt.show()
```

```python
img_eq = exposure.equalize_hist(camera)
```

```python
plt.subplot(1,2,1)
plt.title("Equalize image")
plt.imshow(img_eq,'gray')
plt.subplot(1,2,2)
plt.title("Original image")
plt.imshow(camera,'gray')
```
