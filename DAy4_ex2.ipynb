```python
import matplotlib
import matplotlib.pyplot as plt

import numpy as np

import skimage
from skimage import data
from skimage import exposure
```

```python
im = data.moon()
```

```python
plt.imshow(im,'gray')
```

```python
wt,ht = im.shape
```

```python
print('กว้าง =',wt,' ','ยาว =',ht)
```

```python
im_new = im
thrd = 225
```

```python
for x in range(wt): 
      for y in range(ht): 
                if im[x, y] >= thrd: 
                        im_new[x, y] = 255
                else:
                        im_new[x, y] = 0
plt.imshow(im_new,'gray')
```
